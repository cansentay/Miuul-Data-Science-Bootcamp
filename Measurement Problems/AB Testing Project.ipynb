# **2. Varsayım Kontrolü**


Normallik Varsayımı

Varyans Homojenliği

#**Normallik Varsayımı**

H0: Normal dağılım varsayımı sağlanmaktadır.

H1:..sağlanmamaktadır.

```python
test_stat, pvalue = shapiro(df["Purchase"])
print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

```python
test_stat, pvalue = shapiro(df1["Purchase"])
print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

** P value 0.05ten büyük büyük olduğu için H0 Kabul**

**Yani Normallik dağılımı sağlanmaktadır**


# **Varyans Homojenligi Varsayımı**


H0: Varyanslar Homojendir

H1: Varyanslar Homojen Değildir

```python
test_stat, pvalue = levene(df["Purchase"],
                           df1["Purchase"])
print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

** P value 0.05ten büyük büyük olduğu için H0 Kabul**

**Yani Normallik dağılımı sağlanmaktadır**


# **3 ve 4. Hipotezin Uygulanması**


 1. Varsayımlar sağlanıyorsa bağımsız iki örneklem t testi (parametrik test)

 2. Varsayımlar sağlanmıyorsa mannwhitneyu testi (non-parametrik test)

```python
test_stat, pvalue = ttest_ind(df["Purchase"],
                              df1["Purchase"],
                              equal_var=True)
print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

**P value 0.05ten büyük büyük olduğu için H0 Kabul**

**Yapılan işlemin satın almaya istatiksel olarak bir etkisi olmamıştır**
